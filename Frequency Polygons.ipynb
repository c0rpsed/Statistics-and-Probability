```r
# ggplot2 examples
library(ggplot2) 
```

```r
data = read.csv('BCS3A.csv')
```

```r
data
```

```r
qplot( data$Quiz..15., geom = "freqpoly", bins = 5, xlab='BCS3A Quiz', ylab='Frequency', col=I('darkblue'))
```

```r
qplot( data$Assignment..10.	, geom = "freqpoly", bins = 5, xlab='BCS3A Asignment', ylab='Frequency', col=I('darkblue'))
```

```r
qplot( data$MidTerm..25., geom = "freqpoly", bins = 5, xlab='BCS3A Midterm', ylab='Frequency', col=I('darkblue'))
```

```r
qplot( data$Final..50., geom = "freqpoly", bins = 5, xlab='BCS3A Final', ylab='Frequency', col=I('darkblue'))
```

```r
qplot( data$Total.100., geom = "freqpoly", bins = 5, xlab='BCS3A Total', ylab='Frequency', col=I('darkblue'))
```
